```python
#import libraries
import pandas as pd
import numpy as np
import os
import pickle
```

```python
df_ords_prods_merged=pd.read_pickle(os.path.join(r'C:\Users\jadep\Documents\Career Foundry\Achievement 4\04-2020Instacart Basket Analysis\Data\Prepared Data\ords_prods_merged.pkl'))
```

```python
#create 'max order'
df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merged.head()
```

```python
#create 'loyalty flag'
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#check for 'loyalty flag'
df_ords_prods_merged.head()
```

```python
#average price by customer loyalty
df_ords_prods_merged.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

```python
#create 'avg price'
df_ords_prods_merged['avg_price'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merged.head()
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['avg_price'] <10, 'spending_flag'] = 'Low spender'
df_ords_prods_merged.loc[df_ords_prods_merged['avg_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
#limit dataset for memory
df2=df_ords_prods_merged[:10000000]
```

```python
#create 'spending flag'
df2.loc[df2['avg_price'] <10, 'spending_flag'] = 'Low spender'
df2.loc[df2['avg_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
#check spending flag counts
df2['spending_flag'].value_counts(dropna=False)
```

```python
#create 'order frequency'
df_ords_prods_merged['order_frequency'] = df_ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merged.head()
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['order_frequency'] >20, 'frequency_flag'] = 'Non frequent customer'
df_ords_prods_merged.loc[(df_ords_prods_merged['order_frequency'] <= 20) & (df_ords_prods_merged['order_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
df_ords_prods_merged.loc[df_ords_prods_merged['order_frequency'] < 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#limit dataframe for memory
df2=df_ords_prods_merged[:1000000]
```

```python
df2.loc[df2['order_frequency'] >20, 'frequency_flag'] = 'Non frequent customer'
df2.loc[(df2['order_frequency'] <= 20) & (df2['order_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
df2.loc[df2['order_frequency'] < 10, 'frequency_flag'] = 'Frequent customer'
```
